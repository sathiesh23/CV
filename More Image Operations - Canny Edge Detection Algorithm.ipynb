## **EDGE DETECTIO**N

The Canny edge detector is arguably the most well known and the most used edge detector in all of computer vision and image processing.

**Steps involved in Canny Edge Detection Algorithm**:


1.   Applying Gaussian smoothing to the image to help reduce noise.
2.   Computing the Gx and Gy image gradients using the Sobel kernel.
3.   Applying non-maxima suppression to keep only the local maxima of gradient magnitude pixels that are pointing in the direction of the gradient.
4.   Defining and applying the Tupper and Tlower thresholds for Hysteresis thresholding.

```python
# import the necessary packages
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import argparse
import cv2
import numpy as np
```

```python
from google.colab import files
apj = files.upload()
```

```python
# load the image, convert it to grayscale, and blur it slightly
image = cv2.imread("apj.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# show the original and blurred images
print("Original Image")
cv2_imshow(image)
print("\n")
print("Grayscale Image")
cv2_imshow(gray)
print("\n")
print("Gaussian Blurred Image")
cv2_imshow(blurred)
print("\n")
```

```python
# compute a "wide", "mid-range", and "tight" threshold for the edges
wide = cv2.Canny(blurred, 10, 200)
mid = cv2.Canny(blurred, 50, 150)
tight = cv2.Canny(blurred, 240, 250)

# show the edge maps
print("Wide Edge Map")
cv2_imshow(wide)
print("\n")
print("Mid Edge Map")
cv2_imshow(mid)
print("\n")
print("Tight Edge Map")
cv2_imshow(tight)
```
